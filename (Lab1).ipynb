4.Convert the daily to weekly format and plot it.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("SBI_stock_history.csv", parse_dates=["tradeDate"])


df.set_index("tradeDate", inplace=True)
df.sort_index(inplace=True)


weekly = pd.DataFrame()
weekly["closingPrice"] = df["closingPrice"].resample("W-FRI").last()
weekly["maxPrice"] = df["maxPrice"].resample("W-FRI").max()
weekly["minPrice"] = df["minPrice"].resample("W-FRI").min()
weekly["volume"] = df["volume"].resample("W-FRI").sum()
weekly["amount"] = df["amount"].resample("W-FRI").sum()
weekly["noOfTransactions"] = df["noOfTransactions"].resample("W-FRI").sum()


weekly.dropna(inplace=True)


plt.figure(figsize=(12, 6))
plt.plot(
    weekly.index,
    weekly["closingPrice"],
    color="#1f77b4",
    linewidth=2,
    label="SBI Weekly Closing Price"
)

plt.title("Weekly Closing Price - SBI Bank", fontsize=16, weight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()





```

5.Convert the data to monthly format and plot it.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("SBI_stock_history.csv", parse_dates=["tradeDate"])


df.set_index("tradeDate", inplace=True)
df.sort_index(inplace=True)


monthly = pd.DataFrame()
monthly["closingPrice"] = df["closingPrice"].resample("M").last()
monthly["maxPrice"] = df["maxPrice"].resample("M").max()
monthly["minPrice"] = df["minPrice"].resample("M").min()
monthly["volume"] = df["volume"].resample("M").sum()
monthly["amount"] = df["amount"].resample("M").sum()
monthly["noOfTransactions"] = df["noOfTransactions"].resample("M").sum()


monthly.dropna(inplace=True)


plt.figure(figsize=(12, 6))
plt.plot(
    monthly.index,
    monthly["closingPrice"],
    color="#ff7f0e",
    linewidth=2,
    label="SBI Monthly Closing Price"
)

plt.title("Monthly Closing Price - SBI Bank", fontsize=16, weight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

6.Calculate the following technical indicators from the data and plot them.
a.Moving Average
b.Weighted Moving Average
c.RSI
d.OBV
e.CCI


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("SBI_stock_history.csv", parse_dates=["tradeDate"])
df.set_index("tradeDate", inplace=True)
df.sort_index(inplace=True)
df.dropna(inplace=True)
```

```python
df['MA_20'] = df['closingPrice'].rolling(window=20).mean()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['closingPrice'], label='Closing Price', color='skyblue')
plt.plot(df.index, df['MA_20'], label='20-Day MA', color='orange', linewidth=2)
plt.title("Moving Average (20 Days) - SBI", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def weighted_moving_average(data, window):
    weights = np.arange(1, window + 1)
    return data.rolling(window).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)

df['WMA_20'] = weighted_moving_average(df['closingPrice'], 20)

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['closingPrice'], label='Closing Price', color='gray')
plt.plot(df.index, df['WMA_20'], label='Weighted MA (20)', color='green', linewidth=2)
plt.title("Weighted Moving Average (20 Days) - SBI")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("SBI_stock_history.csv", parse_dates=["tradeDate"])


df.set_index("tradeDate", inplace=True)
df.sort_index(inplace=True)


def compute_rsi(series, period=14):
    delta = series.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df["RSI_14"] = compute_rsi(df["closingPrice"])
df.dropna(subset=["RSI_14"], inplace=True)


plt.figure(figsize=(12, 6))
plt.plot(df.index, df["RSI_14"], label="RSI (14)", color="purple", linewidth=1.5)
plt.axhline(70, color="red", linestyle="--", label="Overbought (70)")
plt.axhline(30, color="green", linestyle="--", label="Oversold (30)")

plt.title("RSI (14 Days) - SBI Bank", fontsize=16, fontweight='bold')
plt.xlabel("Date")
plt.ylabel("RSI")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
df['OBV'] = (np.sign(df['closingPrice'].diff()) * df['volume']).fillna(0).cumsum()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['OBV'], color='darkblue', label='OBV')
plt.title("On-Balance Volume (OBV) - SBI")
plt.xlabel("Date")
plt.ylabel("OBV")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
tp = (df['maxPrice'] + df['minPrice'] + df['closingPrice']) / 3
sma = tp.rolling(20).mean()
mad = tp.rolling(20).apply(lambda x: np.fabs(x - x.mean()).mean())
df['CCI'] = (tp - sma) / (0.015 * mad)

plt.figure(figsize=(12, 5))
plt.plot(df.index, df['CCI'], color='teal', label='CCI (20)')
plt.axhline(100, color='red', linestyle='--')
plt.axhline(-100, color='green', linestyle='--')
plt.title("Commodity Channel Index (CCI) - SBI")
plt.ylabel("CCI")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
